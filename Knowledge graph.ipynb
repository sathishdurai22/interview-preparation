```python
from neo4j import GraphDatabase
system_message="You are an agent designed to retrieve movie data from a Neo4j database."
from langchain_core.messages import AIMessage

class GraphDBRetrieveAgent:
    def __init__(self, graph, system_message=system_message):
        self.graph = graph
        self.system_message = system_message

    def query_graph(self, query, params=None):

        result = self.graph.query(query, params)
        return result

    def retrieve_answer(self, user_query):
        if isinstance(user_query, AIMessage):
            user_query = user_query.content  
        
        # Construct a parameterized Cypher query
        cypher_query = """
        MATCH (n:Movie)  // Use your label 'Movie' here
        WHERE n.title CONTAINS $user_query  // Use 'title' as the search property
        RETURN n.title, n.plot, n.revenue, n.year LIMIT 1;
        """

        # Execute the query with the user query as a parameter
        params = {"user_query": user_query}  
        results = self.query_graph(cypher_query, params)

        for record in results:
            return f"Found: {record['n.name']} - {record['n.description']}"

        return "No results found for your query."

```

```python

agent = GraphDBRetrieveAgent(graph=graph)

user_query = "Toy Story country of production"

response = agent.retrieve_answer(user_query)

print(response)
```
